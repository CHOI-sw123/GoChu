#### 이미지 증식을위한 제너레이터

제너레이터 파라미터 조정 

```python
# 증식 개수 조정 
image_cnt =  35*all_dept.shape[0]  # 증식 데이터 개수 장당35개로 조정

#PC값 조정
pc_cnt = 10 # pc개수
interval = -5


# Generator 파라미터값 세부 조정 
#=============================================================

width_shift_range = 0.3 # 가로움직임
height_shift_range =  0.3 # 세로움직임
shear_range = 0.3 # 기울어짐 정도
zoom_range = 0.3# 확대 비율
fill_mode = "nearest" # 확대 시에 이미지가 깨지는 문제를 보정

#=============================================================

##PCA 시드값은 함수에서 직접 조정
```

```python
#증식
#======================================================================================

#증식을 위한 Reshape
all_dept_re1 = all_dept.reshape(all_dept.shape[0], # 개수
                                all_dept.shape[1], # 가로 픽셀
                                all_dept.shape[2], # 세로 픽셀
                                1)                 # 채널수(흑백)

# PCA 학습 대상 이미지 (== 증식대상 이미지 데이터셋)
tmp = all_dept_re1

# 생성할 이미지 
input_image = tmp.copy()

# 이미지를 생성시켜주는 모델 양식을 생성
train_gen = ImageDataGenerator(
#                               rescale=1.0/255, #rescale은 이미 진행했으니 필요없을듯 
#                               rotation_range=5,  #회전각도
                              width_shift_range=width_shift_range,  
                              height_shift_range=height_shift_range,  
                              shear_range=shear_range,  
                              zoom_range = zoom_range,  
                              fill_mode = fill_mode  
                              )

# 실제로 이미지를 생성
train_generator=train_gen.flow(tmp.copy(),
                               batch_size=1)

for i in range(image_cnt):
    image = train_generator.next()  
    tmp = np.append(tmp.copy(), image, axis=0) 
    
#=======================================================================================

#pca모델 생성
all_dept_PCA_model =PCA(n_components=pc_cnt, # PC개수 
                        random_state=2 # 시드 조정
                       ) 

#학습
tmp_re = tmp.reshape(tmp.shape[0],-1) # PC 설정을 위한 fit용 데이터셋 reshape
all_dept_PCA_model.fit(tmp_re) #증식한 이미지 학습

#주성분 추출
all_dept_re2 = all_dept.reshape(all_dept.shape[0],-1) # 주성분 추출을 위한 trans용 데이터셋 reshape
all_dept_trans = all_dept_PCA_model.transform(all_dept_re2) #비교할 현직자들 필체 주성분 추출
```
