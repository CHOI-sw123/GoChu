```python
def handwriting(image_name,data_set ,model):
    img_num = ["_0", "_1","_2","_3","_4","_5","_6"]
    student_dir = "./image/ver.7/student/"
    teacher_dir = "./image/ver.7/teacher/" #경로설정.
    
    
    
    #필기체 경로를 가져와서 전처리까지 진행하고 비교용 데이터셋까지 만드는 함수
    
                            #경로        #이미지이름
    student0_1 = cv2.imread(student_dir+image_name+img_num[1]+".jpg", cv2.IMREAD_GRAYSCALE)
    student0_2 = cv2.imread(student_dir+image_name+img_num[2]+".jpg", cv2.IMREAD_GRAYSCALE)
    student0_3 = cv2.imread(student_dir+image_name+img_num[3]+".jpg", cv2.IMREAD_GRAYSCALE)
    student0_4 = cv2.imread(student_dir+image_name+img_num[4]+".jpg", cv2.IMREAD_GRAYSCALE)
    student0_5 = cv2.imread(student_dir+image_name+img_num[5]+".jpg", cv2.IMREAD_GRAYSCALE)
    student0_6 = cv2.imread(student_dir+image_name+img_num[6]+".jpg", cv2.IMREAD_GRAYSCALE)

    scaled_student0_1 = preprocessing.scale(student0_1.T)
    scaled_student0_2 = preprocessing.scale(student0_2.T)
    scaled_student0_3 = preprocessing.scale(student0_3.T)
    scaled_student0_4 = preprocessing.scale(student0_4.T)
    scaled_student0_5 = preprocessing.scale(student0_5.T)
    scaled_student0_6 = preprocessing.scale(student0_6.T)

    #데이터 셋 만들기
    #student
    student0 = np.array([scaled_student0_1,scaled_student0_2,scaled_student0_3,scaled_student0_4,scaled_student0_5,scaled_student0_6])
    student0_re = student0.reshape(student0.shape[0],-1)

    #학생 transform
    student0_trans = model.transform(student0_re)
    
    return similarity_dept(student0_trans, data_set, model)
```
